```python
import matplotlib.pyplot as plt
```

```python
# sorting first 10 countries by population in 1960, 2000 and 2022
df_sorted1960 = df.sort_values(by='1960', ascending=False).head(10)
df_sorted2000 = df.sort_values(by='2000', ascending=False).head(10)
df_sorted2022 = df.sort_values(by='2022', ascending=False).head(10)
```

```python
#Bar Chart for the year 1960
plt.figure(figsize=(10, 6))
plt.barh(df_sorted1960['Country Name'], df_sorted1960['1960'], color='green')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 1960')
plt.gca().invert_yaxis() 
plt.show()
```

```python
#Bar Chart for the year 2000
plt.figure(figsize=(10, 6))
plt.barh(df_sorted2000['Country Name'], df_sorted2000['2000'], color='brown')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 2000')
plt.gca().invert_yaxis() 
plt.show()
```

```python
#Bar Chart for the year 2022
plt.figure(figsize=(10, 6))
plt.barh(df_sorted2022['Country Name'], df_sorted2022['2022'], color='skyblue')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 2022')
plt.gca().invert_yaxis() 
plt.show()
```

```python
df_sorted1960 = df.sort_values(by='1960', ascending=False)
df_sorted2000 = df.sort_values(by='2000', ascending=False)
df_sorted2022 = df.sort_values(by='2022', ascending=False)
```

```python
# extracting last 10 countries from bottom by population
df_last1960 = df_sorted1960.tail(10)
df_last2000 = df_sorted2000.tail(10)
df_last2022 = df_sorted2022.tail(10)
```

```python
#Bar Chart for the year 1960
plt.figure(figsize=(10, 6))
plt.barh(df_last1960['Country Name'], df_last1960['1960'], color='green')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 1960')
plt.gca()
plt.show()
```

```python
#Bar Chart for the year 2000
plt.figure(figsize=(10, 6))
plt.barh(df_last2000['Country Name'], df_last2000['2000'], color='brown')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 2000')
plt.gca()
plt.show()
```

```python
#Bar Chart for the year 2022
plt.figure(figsize=(10, 6))
plt.barh(df_last2022['Country Name'], df_last2022['2022'], color='skyblue')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 2022')
plt.gca()
plt.show()
```

```python
# Bar Chart for Income Distributions by Regions
income_by_region= pd.crosstab(mdata.IncomeGroup, mdata.Region)
plt.figure(figsize=(10, 6))
income_by_region.plot(kind='bar')
plt.xlabel("Income Group")
plt.ylabel("Number of Countries in the Regions")
plt.title("Income Distribution by Regions")
plt.legend(title='Regions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for population distribution in India from 1960-2022
India_data = df[df['Country Code'] == 'IND']
years = [str(year) for year in range(1960, 2023)]  # Years from 1960 to 2022
population = India_data[years].values.flatten()
plt.figure(figsize=(10, 6))
plt.hist(population, bins=15, edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution in India (1960-2022)')
plt.show()
#distribution of population sizes within India during this time period.
```
